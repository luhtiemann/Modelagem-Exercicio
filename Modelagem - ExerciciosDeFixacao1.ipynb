3. Usando o Python e as fun√ß√µes trigonom√©tricas dispon√≠veis nesta linguagem de programa√ß√£o e suas bibliotecas crie uma fun√ß√£o que receba um √¢ngulo, em graus, e plote um circulo trigonom√©trico de raio 1, centrado na origem destacando este √¢ngulo, seu seno, seu cosseno e sua tangente.

```python
def circTrig():

  plt.close('all') #s√≥ para ter certeza

  ang = float(input("Insira o √¢ngulo desejado: "))

  #transforma para radianos
  s = math.sin(math.radians(ang)) 
  c = math.cos(math.radians(ang))
  t = math.tan(math.radians(ang))

  print(f"√Çngulo: {ang}\nSeno: {s:.2f}\nCosseno: {c:.2f}\nTangente: {t:.2f}.")

  # cria uma figura de 18x7 polegadas
  plt.figure(figsize = (18, 7))

  #origem
  plt.plot(0,0, color = 'black', marker = 'o')

  #circulo
  angles = linspace(0*pi, 2*pi, 100 )
  xs = cos(angles)
  ys = sin(angles)
  plt.plot(xs, ys, color = 'black')

  #raio
  plt.plot(1, 0, marker = 'o', color = 'green')
  plt.plot([0, 1], [0, 0], color = 'black')

  #arco
  arc_angles = linspace(0, math.radians(ang), 20)
  arc_xs = cos(arc_angles)
  arc_ys = sin(arc_angles)
  plt.plot(arc_xs, arc_ys, color = 'green', lw = 3)

  #outro raio
  plt.plot(cos(math.radians(ang)), sin(math.radians(ang)), marker = 'o', color = 'green')
  plt.plot([0, cos(math.radians(ang))], [0, sin(math.radians(ang))], color = "green", label="√¢ngulo")

  #√¢ngulo
  r1 = 0.25
  arc_angles = linspace(0 * pi, math.radians(ang), 20)
  arc_xs = r1 * cos(arc_angles)
  arc_ys = r1 * sin(arc_angles)
  plt.plot(arc_xs, arc_ys, color = 'green', lw = 3)

  #seno
  plt.plot(0, s, marker = 'o', color = 'purple')
  plt.plot([0, 0], [0, s], color = 'purple', label = "seno")

  #cosseno
  plt.plot(c, 0, marker = 'o', color = 'blue')
  plt.plot([c, 0], [0, 0], color = 'blue', label = "cosseno")

  #tangente
  plt.plot(1, 0, marker = 'o', color = 'red')
  plt.plot(1, t, marker = 'o', color = 'red')
  plt.plot([1,1], [0, t], color = 'red', label = "tangente")

  #limites
  plt.xlim(-2, 2)
  plt.ylim(-2, 2)
  plt.gca().set_aspect('equal')
  plt.legend(loc='upper left')

  plt.show()

circTrig()
```

```python
from pickle import TRUE

def Periodos():
  plt.close('all') #s√≥ para ter certeza
  plt.figure(figsize = (10, 5)) #tamanho da figura

  while TRUE:
    n = int(input("Insira o um n√∫mero de per√≠odos completos, sendo um inteiro maior ou igual a 1 e menor ou igual a 5: ")) #input
    if n>=1 and n<=5: #limite do input

      x = np.linspace(0, 2*n*pi, 300) #cria uma sequencia de n√∫meros uniformemente espa√ßados entre os limites dados
      c = np.cos(x) #calcula seno, cosseno, tangente
      s = np.sin(x)
      t = np.tan(x) 

      t[:-1][np.diff(t) < 0] = np.nan #how to plot tan without connecting line

      plt.plot(x, s, color="purple", label = "seno") #plotar, sequencia de numeros, seno, cor, legenda
      plt.plot(x, c, color="blue", label = "cosseno")
      plt.plot(x, t, color="red", label = "tangente") 
      plt.legend(loc='upper left')

      plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$']) #valores laterais

      plt.xlim(0, 2*n*pi) #limites
      plt.ylim(-1.5, 1.5)  

      plt.show()

      break

    else:
      print("Insira um n√∫mero entre 1 e 5")

  
Periodos()
```
